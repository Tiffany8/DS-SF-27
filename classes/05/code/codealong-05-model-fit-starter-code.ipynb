## Your first Machine Learning Models!

### Machine Learning Model #1 | `SalePrice` as a function of `M1`

```python
# TODO
X = df[ ['M1'] ]
y = df.SalePrice

model = smf.OLS(y, X).fit()
```

```python
model.summary()
```

### Machine Learning Model #2 | `SalePrice` as a function of `M2`

```python
# TODO
X_m2 = df[ ['M2'] ]
y_m2 = df.SalePrice

model_m2 = smf.OLS(y_m2, X_m2).fit()
model_m2.summary()
```
